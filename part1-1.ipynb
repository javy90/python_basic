```python
#판다스의 시리즈
import pandas as pd

a_series = pd.Series({'a':10, 'b':20, 'c':30, 'd':40}) #딕셔너리의 키가 인덱스명, 값이 value로 변환
print(a_series)
print(type(a_series))
```

```python
# 시리즈의 데이터 선택 : 시리즈명[] -> 인덱스명 또는 인덱스 첨자로 불러옴
print(a_series.index, a_series.values,'\n', a_series.value_counts())
print(a_series['a'],a_series[0])
print(a_series[[0,1,2]]) # 여러개의 값을 검색 [[안에 넣어서]]
print(a_series[0:2]) # 숫자인덱스 start:end -> end 미포함
print(a_series['a':'c']) # 인덱스명 start:end -> end 포함
```

```python
#데이터프레임 : 여러개의 시리즈가 모여서 만들어짐
#딕셔너리의 키는 컬럼명(시리즈 이름이 됨) 행데이터는 row 데이터 새로 칼럼이름
#pandas.DataFrame(딕셔너리,[index = 행이름], [columns= 열이름])
dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9],'c3':[10,11,12]}
df = pd.DataFrame(dict_data)#데이터프레임 생성
print(df)
print(df.columns, df.index)# 컬럼명, 인덱스명 검색
df.index = ['a','b','c']#새로운 인덱스명 부여
print(df)
print(df.columns, df.index)
df.columns = ['a1','a2','a3','a4']
print(df)
print(df.columns, df.index)
df
```

```python
# 데이터프레임 : 2차원 배열 (시리즈가 열로 묶인 배열)
# 인덱스 접근방법은 : df.loc[인덱스명], 값이 여러개이면 []안에!
# 칼럽 접근 방법은 : df[컬럼명]
# 원소 접근방법 : df.loc[인덱스명,컬럼명], 
#                 df.iloc[인덱스첨자,컬럼첨자]
# 인덱스 생성 : df.index = [인덱스명, ...]
# 특정 컬럼은 인덱스로 : df.set_index(컬럼명, inplace=True)
# 인덱스 초기화 : df.reset_index(inplace=True)
# 인덱스 재배열 : df.reindex([인덱스명,...]) 
# 인덱스 삭제 : df.drop(index=[인덱스명])
# 칼럼 삭제 : df.drop(columns=[칼럼명], axis=1)
# 인덱스 정렬 : df.sort_index()
# 값으로 정렬 : df.sort_values(by=[],ascending=False)
```

```python
#시리즈 연산 :시리즈 + 사칙연산 +숫자
sr1 = pd.Series({'국어':100,'수학':90,"영어":90})
print(sr1)
#  각 점수에 100을 더한 값을 출력
print(sr1+100)
#  각 점수에 100을 나눈 값을 출력
print(sr1/100)
```

```python
##시리즈 연산 :시리즈 + 사칙연산 + 시리즈
sr1 = pd.Series({'국어':100,'수학':90,"영어":90})
sr2 = pd.Series({'영어':100,'국어':90,"수학":90,'과학':90}) #인덱스가 없는 경우 nan
print(sr1+sr2)

print(sr1.add(sr2,fill_value=0))
#인덱스가 없는 경우 연산메서드 중 fill_value를 넣어 nan 값 해결하고 진행해야함
```
