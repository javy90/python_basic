```python
# 문 titanic 데이터셋을 불러와서
#1.나이가 10세 미만인 승객의 class, 성별, 나이, 생존여부를 출력
#2. 클래스가 1등급인 승객의 나이, fare, 클래스, 생존여부를 출력
#3. 성별이 남성이고 20대에서 40대 미만인 승객의 모든 정보 출력
import seaborn as sns
df = sns.load_dataset('titanic')
df_1 = df.loc[df.age<10,['class','age','sex','alive']]
print(df_1)
df_2 = df.loc[df['class'] == 'First',['age','fare','class','alive']]
print(df_2)
df_3 = df.loc[(df.sex == 'male')&(df.age>=20) & (df.age<40),:]
print(df_3)
df.info()
```
