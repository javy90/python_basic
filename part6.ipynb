```python
def add_10(n):
    return n+10
def add_two_obj(a,b):
    return a+b
print(add_10(20)) #30
# print(n) #n은 add_10 함수가 종료되면 사라짐
print(add_two_obj(10,10)) # 20
```

```python
# 시리즈 객체의 각 원소에 함수를 매핑 .apply(매핑함수)
sr1 = df['age'].apply(add_10) #add_10(n), n <- df.age
print(df['age'].head(3));print()
print(sr1.head(3))

#시리즈 객체와 숫자를 적용한 함수 add_two obj(a,b)
sr2 = df['age'].apply(add_two_obj, b=10) # a <- df.age
print(sr2.head(3))

#매핑함수를 lamda 함수로 정의해서 사용
sr3 = df['age'].apply(lambda x: add_10(x))#x <- df['age']각각의 원소
print(sr3.head(3))
```

```python
#apply 함수를 데이터 프레임에 적용
#.apply(매핑함수, axis=0)->컬럼을 시리즈로

def missing_value(series):
    return series.isnull()
result = df.apply(missing_value)
result
```

```python
#데이터프레임의 행 단위로 원소에 함수를 적용
df['ten']=10
df.head(3)

df['result'] = df.apply(lambda x: add_two_obj(x['age'],x['ten']),
                        axis=1)
df.head(3)
```

```python
#pipe() : 데이터프레임 객체를 함수에 매칭
df= df.loc[:,['age','fare']]
df.head(3)

def missing_value(x):
    return x.isnull()
def missing_count(x):
    return missing_value(x).sum()
def total_missing_number(x):
    return missing_count(x).sum()

result_df = df.pipe(missing_value)
print(result_df.head(3))
print(type(result_df))

result_sr = df.pipe(missing_count)
print(result_sr.head(3))
print(type(result_sr)) #series

result_total = df.pipe(total_missing_number)
print(result_total,type(result_total))
```

```python
#merge를 활용하여 데이터프레임 합치기
df1 = pd.read_excel('./dataset/stock_price.xlsx')
df2 = pd.read_excel('./dataset/stock_valuation.xlsx')

print(df1.columns) #<=id,stock_name
print(df2.columns) #<=id,name
```

```python
#기본값 on: 제일 왼쪽칼럼명을 index로써 기준
#how='inner' 같은행열만 표현
#양쪽의 데이터프레임의 'id'컬럼이 같은 자료만 병합
pd.merge(df1,df2) 
#left인 df1을 기준으로 병합한다.
#df1은 모두 출력, df2 없으면 Nan
pd.merge(df1,df2,how='left')
#right인 df2을 기준으로 병합한다.
#df2은 모두 출력, df1 없으면 Nan
pd.merge(df1,df2,how='right')
#outer : df1, df2모두출력 없으면 Nan
pd.merge(df1,df2,how='outer',on='id')# on='key' 양쪽에 있는 컬럼명을 넣음
#df1의 'stock_name',df2의 'name' 컬럼의 값이 같은 자료를 병한
result=pd.merge(df1,df2,how='left',left_on='stock_name',right_on='name')

r_data = result.loc[:,['id_x','id_y','name','value','price','eps']]
r_data
#df1의 price가 50000미만인 자료와 df2을 결합
pd.merge(df1[df1.price < 50000],df2)
```

```python
#['survived','age','sex','class','fare'] : 컬럼만 분리
df = titanic.loc[:,['age','sex','class','fare','survived']] 
# 컬럼명이 나열 순서대로 만들어짐
# class를 기준으로 분할
grouped = df.groupby('class')

for key, group in grouped:
    print("key : {}, 승객수 : {}".format(key, len(group)))
    print(group.head());print()
```

```python
#그룹연산 메소드 데이터 집계(aggreagation)
#class로 그룹을 지어 fare 컬럼만 std()표준편차 구함
grouped = df.groupby('class')
print(grouped.fare.std().round(2)) #각 그룹의 원하는 컬럼 추출 후 함수적용 
#1회만 실행
grouped.std().fare# 각 그룹에 함주적요 후 원하는 컬럼추출
# 그룹이 3개일 시 3회실행
```

```python
#집계연산을 처리 -> 사용자 함수를 그룹객체에 적용
df = titanic.loc[:,['age','class','fare','survived']] 
grouped = df.groupby('class')
#group 객체.agg(매핑 함수) 함수가 여러개일 겨우 list
def max_min(x):
    return x.max()-x.min()
#grouped 객체에 max_min함수를 적용
grouped.agg(max_min)
# grouped.max()
# grouped.min()

#모든 열에 여러 함수를 매핑 : 그룹객체.agg([함수1, 함수2,...])
#각 열마다 다른 함수를 매핑 : 그룹객체.agg({열1:함수1,열2:함수2,...})
agg_all = grouped.agg(['max','min']) # 함수 str 값으로
print(agg_all)

agg_sep = grouped.agg({'age':'mean','fare':['min','max']})
agg_sep
```

```python
#agg(),transform() = 각 그룹별 데이터에 함수적용, 각 원소에 함수적용
#그룹객체.transform(매핑함수) : 각 그룹의 행에 대한 연산
age_mean = grouped.age.mean() # 그룹별 'age'컬럼의 평균 
type(age_mean)
print(age_mean) #그룹의 key가 인덱스로 설정되어 있음
print('------')
age_std = grouped.age.std() # 그룹별 'age'컬럼의 표준편차
print(age_std)
print('------')

# 그룹 객체의 age 컬럼에 대해 zscore 계산하여 출력
for key, group in grouped.age:
    group_zscore = (group - age_mean.loc[key]) / age_std.loc[key]
    print("origin : ", key)
    print('------')
    print(group_zscore.head(3))
print('---위 로직을 아래처럼 간편하게---')  
def z_score(x):
    return((x-x.mean())/x.std())
age_zscore = grouped.age.transform(z_score)
age_zscore.loc[0:9]
```
